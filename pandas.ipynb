```python
df1=pd.DataFrame({'1key':['foo','bar','baz','foo'],
                  'value':[1,2,3,5]
                  })

df2=pd.DataFrame({'rkey':['foo','bar','baz','foo'],
                  'value':[5,6,7,8]
                  })

df1.merge(df2,left_on='1key',right_on='rkey',suffixes=("_left","_right"))
```

```python
df1=pd.DataFrame({'1key':['foo','bar','baz','foo'],
                  'value':[1,2,3,5]
                  })

df2=pd.DataFrame({'rkey':['foo','bar','baz'],
                  'value':[5,6,7]
                  })

df1.join(df2,lsuffix='_caller',rsuffix='other')
```
