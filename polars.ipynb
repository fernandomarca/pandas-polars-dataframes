```python
import polars as pl

df=pl.read_csv("ordf.csv")
df
```

```python
df.select(["a","a_or_b"]).sum()
```

```python
df.select("^a|b$")
```

```python
df.select([pl.all(),
          pl.all().reverse().suffix("_reverse")]
          )
```

```python
df.select([pl.all(),
          pl.all().sum().suffix("_sum")]
          )
```

```python
df = pl.DataFrame(
    {
        "A": [1, 2, 3, 4, 5],
        "fruits": ["banana", "banana", "apple", "apple", "banana"],
        "B": [5, 4, 3, 2, 1],
        "cars": ["beetle", "audi", "beetle", "beetle", "beetle"],
        "optional": [28, 300, None, 2, -30],
    }
)
df
```

```python
predicate = pl.col("fruits").str.contains("^b.*")
df.select([predicate])
```

```python
df.filter(predicate)
```

```python
df.filter(pl.col("B")<5)
```
